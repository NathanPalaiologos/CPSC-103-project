# Overview

### What problem is this solving? 

- How much does the overall Canadian CPI change monthly and how does this change compare to two major components of CPI shelter and food which are fundamental necessities? 

### Where is the information from? 

Canadian CPI cleaned.csv

reference Statistics Canada: https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=1810000401

### Why is it interesting? 

- Interest in economics and post covid rising food and rent/housing prices broad increase across Canada.

- CPI is important in calibrating monetary policy, making economic decisions, and gauging economic health. Moreover, this analysis provides insights into cost of living trends, and consumer spending patterns. 

# Design Choices

### What design choices were made, and why were these design choices made? 

- How to design data recipes: Deciding on Enumeration, Compound, Arbitrary-Sized.

We chose Mapping Dictionary instead of Enumeration as a clean and concise way to map months as integers, adding and removing months only require changing the dictionary not requiring code logic change. The final project changed to a simple atomic monthly string  returning the corresponding integer to be more consistent with course material. 

- Visual design choices: Deciding on a line chart, bar graph, or scatterplot and reason for our choice. 
- Our goal for visuals: able to visually show trends and changes. Able to visually see the relationship between food and housing and CPI.
The line chart show changing trends over time including major shifts in the data. Relationship between variables can be visually displayed with one line each for CPI, food, and shelter on same line graph. Scatterplot too many overlapping points. Bar graph more difficult to see relationship between three variables.

### Describe how you solved the problem. Provide visuals if possible, and definitely provide the graph or chart(s) that your program created. 

We solved the problem by (refer to final project calculations):

- Filtered for CPI, Food and Housing monthly data and plotted on line graph Canadian CPI (Base Year 2002) vs. Time. 
- To find the monthly change in CPI or one of its 8 components, we will subtract the difference between monthly CPI divided by the prior month's CPI. (Feb 2013 CPI - Jan 2013 CPI)/Jan 2013 CPI. We plotted on line graph Canadian CPI Monthly Change (Base Year 2002) vs. Time.
